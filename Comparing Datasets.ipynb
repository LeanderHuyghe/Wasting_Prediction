# Load and view our current datasets

# Compare the current datasets

## Comparing baseline against semiyearly

```python
compare_sy = datacompy.Compare(df_baseline,df_sy,
                            join_columns="date",
                            df1_name="baseline_data",
                            df2_name="semiyearly_data")
print(compare_sy.report())
```

## Comparing baseline against monthly

```python
compare_mon = datacompy.Compare(df_baseline,df_mon,
                            join_columns="date",
                            df1_name="baseline_data",
                            df2_name="monthly_data")
print(compare_mon.report())
```

## Comparing monthly against semiyearly

```python
compare_mon_sy = datacompy.Compare(df_mon,df_sy,
                            join_columns="date",
                            df1_name="monthly_data",
                            df2_name="semiyearly_data")
print(compare_mon_sy.report())
```
