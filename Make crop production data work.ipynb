```python
import pandas as pd
import numpy as np

semi_df = pd.read_csv('semiyearly_chosen_columns.csv', parse_dates=['date'])
production_df = pd.read_csv('production.csv', parse_dates=['date'])
```

```python
# create a subset that matches our requirements for the final df
production_df = production_df[production_df['date'] >= '2017-01-01']
production_df = production_df[production_df['district'] != 'Hagar' ]  # drop Hagar because it doesn't match any other district
                                                                      # according to the internet it's a city in Lower Juba(region), however all the districts in Lower Juba already appear in the production dataset
production_df = production_df[['date', 'district', 'Cowpea', 'Maize', 'Sorghum']]
# correct the names od the districts in production_df
production_df = production_df.replace({'district': {'Baydhaba': 'Baidoa', 'Ceel Bur': 'Ceel Buur', 'El berde': 'Ceel Barde', 'Garbaharey/Burdhuubo': 'Garbahaarey', 'Wanla weyn': 'Wanla Weyn'}})
production_df['date'] = production_df['date'].apply(lambda x: x.strftime('%Y-%m-%d'))
production_df
```
